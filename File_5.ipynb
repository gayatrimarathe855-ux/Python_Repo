```python
n = int(input("Enter a number: "))

binary = ""

while n > 0:
    r = n % 2
    binary = str(r) + binary
    n = int(n / 2)

while len(binary) < 4:
    binary = "0" + binary

print(binary)
```

```python
start = int(input("Enter start: "))
end = int(input("Enter end: "))

for i in range(start, end + 1, 2):
    if i < end - 1:
        print(i, end="_")
    else:
        print(i)
```
