```python
n = 5

for i in range(n):
    for j in range(n):
        print("*", end=" ")
    print()
```

```python
n = 5

for i in range(1, n + 1):
    print("* " * i)
```

```python
n = 5

for i in range(n, 0, -1):
    print("* " * i)
```

```python
n = 5

for i in range(n):
    print(" " * (n - i - 1) + "* " * (i + 1))
```

```python
n = 5

for i in range(1, n + 1):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()
```

```python
n = 5
num = 1

for i in range(1, n + 1):
    for j in range(i):
        print(num, end=" ")
        num += 1
    print()
```

```python
n = 5

# Upper part
for i in range(1, n + 1):
    print(" " * (n - i) + "* " * i)

# Lower part
for i in range(n - 1, 0, -1):
    print(" " * (n - i) + "* " * i)
```

```python
n = 5

for i in range(n):
    for j in range(n):
        if i == 0 or i == n - 1 or j == 0 or j == n - 1:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```

```python
n = 5

for i in range(n):
    ch = 'A'
    for j in range(i + 1):
        print(ch, end=" ")
        ch = chr(ord(ch) + 1)
    print()
```
