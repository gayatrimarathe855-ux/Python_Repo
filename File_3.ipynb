```python
val1 = int(input("Enetr the first number"))
val2 = int(input("Enter the second number"))

lcm = val1

while lcm % val2 != 0:
   lcm = lcm + val1

print("LCM =", lcm)
```

```python
val = int(input("Enter first number: "))
val2 = int(input("Enter second number: "))

i = 1
hcf = 1

while i <= val and i <= val2:
    if val % i == 0 and val2 % i == 0:
        hcf = i
    i = i + 1

print("HCF =", hcf)
```
