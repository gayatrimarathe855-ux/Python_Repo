```python
import numpy as np

values = np.random.binomial(1, 0.5, 10)
print("Generated Values:", values)
working = np.sum(values)
print("Working Bulbs:", working)
```

```python
import numpy as np

heads = np.random.binomial(5, 0.5, 5)
print("Number of heads in each experiment:")
print(heads)
```

```python
import numpy as np

arrivals = np.random.poisson(2, 10)
print("Customer Arrivals:", arrivals)
print("Highest Customers in an Hour:", np.max(arrivals))
```

```python
import numpy as np

values = np.random.geometric(0.5, 10)
print("Generated Values:", values)
print("Maximum Attempts:", np.max(values))
```

```python
import numpy as np

print("Bernoulli Distribution:")
print(np.random.binomial(1, 0.5, 5))

print("\nBinomial Distribution:")
print(np.random.binomial(5, 0.5, 5))

print("\nGeometric Distribution:")
print(np.random.geometric(0.5, 5))

print("\nPoisson Distribution:")
print(np.random.poisson(2, 5))
```
