```python
n1 = int(input("Enter first number: "))
n2 = int(input("Enter second number: "))

small = n1

if n2 < n1:
    small = n2

for i in range(small, 0, -1):
    if n1 % i == 0 and n2 % i == 0:
        print("HCF =", i)
```

```python
s = input("Enter a string: ")

n = len(s)

for i in range(n):
    print("*" * i + s[i] + "*" * (n - i - 1))
```
